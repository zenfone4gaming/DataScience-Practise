```python
import pandas as pd, numpy as np
```

```python
tips = pd.read_csv('D:\Ryzen1700\Study\ML\Dataset\pydata-book-2nd-edition\pydata-book-2nd-edition\examples\\tips.csv')
```

```python
tips['tip_pct'] = tips['tip'] / tips['total_bill']
```

```python
tips.pivot_table(index=['day', 'smoker'])
```

```python
tips
```

```python
tips.pivot_table(['tip_pct', 'size'], index=['time', 'day'], columns='smoker')
```

```python

tips.pivot_table(['tip_pct', 'size'], index=['time', 'day'], columns='smoker', margins=True)
```

```python
tips.pivot_table(['tip_pct'], index=['time', 'smoker'], columns='day', margins=True, aggfunc=len)
```

```python
tips.pivot_table(['tip_pct'], index=['time','size', 'smoker'], columns='day', margins=True, aggfunc='mean', fill_value=0)
```

# Cross-Tabulations: Crosstab

A cross tabulation is a special case of a pivot table that computes group frequencies. Here is an example:

```python
pd.crosstab([tips.time, tips.day], tips.smoker, margins=True)
```
