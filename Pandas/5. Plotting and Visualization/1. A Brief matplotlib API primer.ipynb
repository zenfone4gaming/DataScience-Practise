# Adjusting the spacing around subplots

```python
fig, axes = plt.subplots(2,2, sharex=True, sharey=True)
```

```python
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```

# Annotations and Drawing on subplot

```python
from datetime import datetime
import pandas as pd
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
data = pd.read_csv('D:\Ryzen1700\Study\ML\Dataset\sandp500.csv', index_col=0, parse_dates=True)
data
```

```python
spx = data[['Adj Close']]
```

```python
spx.plot(ax=ax, style='k-')
```

```python
crisis_data = [
    (datetime(2020, 2, 1), 'Peak of bull market'),
    (datetime(2020, 3, 3), 'Bear Stearns Fails'),
    (datetime(2020, 3, 31), 'Lehman Bankruptcy'),
]
```

```python
for date, label in crisis_data:
    ax.annotate(label, 
                xy=(date, spx.asof(date)+100), 
                xytext=(date, spx.asof(date)+225),
                arrowprops=dict(facecolor='black', headwidth=4, width=2, headlength=4),
                horizontalalignment='left', 
                verticalalignment='top'
               )
```

```python
ax.set_ylim([2200, 4000])
```
