```python
%matplotlib notebook
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
tips = pd.read_csv("D:\Ryzen1700\Study\ML\Dataset\pydata-book-2nd-edition\pydata-book-2nd-edition\examples\\tips.csv")
party_counts = pd.crosstab(tips['day'], tips['size'])
tips['tip_pct']  = tips['tip'] / (tips['total_bill'] - tips['tip'])
```

```python
sns.factorplot(x='day', y='tip_pct', hue='time', col='smoker', kind='bar', data=tips[tips.tip_pct < 1])
```

```python
sns.factorplot(x='day', y='tip_pct', hue='time', col='smoker', row='time', kind='bar', data=tips[tips.tip_pct < 1])
```
