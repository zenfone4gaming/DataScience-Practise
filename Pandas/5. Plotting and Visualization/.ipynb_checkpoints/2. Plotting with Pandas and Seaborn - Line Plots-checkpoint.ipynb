```python
%matplotlib notebook
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
s = pd.Series(np.random.randn(10).cumsum(), index= np.arange(0, 100, 10))
```

```python
s.plot()
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
                 columns=['A', 'B', 'C', 'D'],
                  index=np.arange(0, 100, 10)
                 )
df.plot()
```
