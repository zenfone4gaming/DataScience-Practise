```python
import pandas as pd
```

```python
#make display smaller
pd.options.display.max_rows = 10
names1880 = pd.read_table('D:\Ryzen1700\Study\ML\Dataset\pydata-book-2nd-edition\pydata-book-2nd-edition\datasets\\babynames\yob1880.txt', sep=',', names=['name','sex','births'] )
names1880
```

```python
names1880.groupby('sex')['births'].sum()
```

```python
years = range(1880, 2011)
pieces = []
columns = ['name', 'sex', 'births']
for year in years:
    path = 'D:\Ryzen1700\Study\ML\Dataset\pydata-book-2nd-edition\pydata-book-2nd-edition\datasets\\babynames\yob%d.txt' % year
    frame = pd.read_csv(path, names=columns)
    frame['year'] = year
    pieces.append(frame)
names = pd.concat(pieces, ignore_index=True)
names
```

```python
total_births = names.pivot_table('births', index='year', columns='sex', aggfunc=sum)
total_births.tail()
```

```python
%matplotlib notebook
total_births.plot(title='Total births by sex and year')
```

```python
def add_prop(group):
    group['prop'] = group.births/group.births.sum()
    return group
names = names.groupby(['year', 'sex']).apply(add_prop)
names
```

```python
names.groupby(['year', 'sex']).prop.sum()
```

```python
#top 1000 names for each year/sex combination
```

```python
def get_top1000(group):
    return group.sort_values(by='births', ascending=False)[:1000]
```

```python
grouped = names.groupby(['year', 'sex'])
top1000 = grouped.apply(get_top1000)
top1000.reset_index(inplace=True, drop=True)#Drop the group index
top1000
```

# Analyzing Naming Trends

```python
boys = top1000[top1000 == 'M']
girls = top1000[top1000 == 'F']
total_births = top1000.pivot_table('births', index='year', columns='name', aggfunc=sum)
total_births
```

```python
subset = total_births[['John', 'Harry', 'Mary', 'Marilyn']]
subset.plot(subplots=True, figsize=(12,10), grid=False, title="Number of births per year")
```

# Measuring the increase in naming diversity

```python
import numpy as np
```

```python
table = top1000.pivot_table('prop', index='year', columns='sex', aggfunc=sum)
table.plot(title='Sum of tabel1000.prop by year and sex', yticks=np.linspace(0,1.2,13), xticks=range(1880, 2020, 10))
```

```python
boys = top1000[top1000['sex']== 'M']
girls = top1000[top1000['sex'] == 'F']
df = boys[boys.year == 2010]
df
```

```python
prop_cumsum = df.sort_values(by='prop', ascending=False).prop.cumsum()
prop_cumsum[:10]
```

```python
prop_cumsum.values.searchsorted(0.5)
```

```python
df = boys[boys.year == 1900]
in1900 = df.sort_values(by='prop', ascending=False).prop.cumsum()
in1900.values.searchsorted(0.5) +1
```

```python
def get_quantile_count(group, q=0.5):
    group = group.sort_values(by='prop', ascending=False)
    return group.prop.cumsum().values.searchsorted(q)+1

diversity = top1000.groupby(['year', 'sex']).apply(get_quantile_count)
diversity
```

```python
diversity = diversity.unstack('sex')
diversity
```

```python
diversity.plot(title="Number of popular names in top 50%")
```

# The "last letter" revolution

```python
#extract last letter from name column
get_last_letter = lambda x: x[-1]
last_letters = names.name.map(get_last_letter)
last_letters
```

```python
names
last_letters.name = 'last_letter'
table = names.pivot_table('births', index=last_letters, columns=['sex','year'], aggfunc=sum)
table
```

```python
subtable = table.reindex(columns=[1920,1960,2010], level='year')
subtable.head()
```

```python
subtable.sum()
```

```python
letter_prop = subtable/subtable.sum()
letter_prop
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 1, figsize=(10, 8))
letter_prop['M'].plot(kind='bar', rot=0, ax=axes[0], title='Male')
#letter_prop['F'].plot(kind='bar', rot=0, ax=axes[0], title='Female',legend=False)
```

```python
letter_prop['F'].plot(kind='bar', rot=0, ax=axes[0], title='Female',legend=False)
```
