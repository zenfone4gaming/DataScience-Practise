```python
chunker =  pd.read_csv('data/Episodes.csv', chunksize =6)
```

```python
tot = pd.Series([])
```

```python
for piece in chunker:
    tot = tot.add(piece['heroes'].value_counts(), fill_value=0)
```

```python
tot = tot.sort_values(ascending=False)
```

# Working With Delimited Formats 

```python
import csv
```

```python
csv.reader?
```

```python
class my_dialect(csv.Dialect):
    lineterminator = '\n'
    delimiter = ','
    quotechar = '"'
    quoting = csv.QUOTE_MINIMAL
```

```python
with open('heroes.csv') as f:
    lines = list(csv.reader(f, dialect=my_dialect))
```

```python
header , value = lines[0], lines[1:]
```

```python
header
```

```python
zip(*value)
```

```python
data_dict = {h : v for h, v in zip(header, zip(*value))}
```
